```python
import pandas as pd
import pytz

data = pd.read_csv("Data/acndata_sessions 2019_11 till 2020_5 caltech.csv")
data.shape[1]
df = data.drop(data.columns[19:93], axis=1)
df = df.drop(df.columns[[0,1,6,9,10]], axis = 1)

# Define time zones
gmt_tz = pytz.timezone("GMT")  # UTC/GMT
pacific_tz = pytz.timezone("America/Los_Angeles")  # PST/PDT

# Convert relevant datetime columns
time_columns = ["connectionTime", "userInputs/0/modifiedAt", "userInputs/0/requestedDeparture"]  # Adjust based on your dataset

for col in time_columns:
    df[col] = pd.to_datetime(df[col], errors="coerce")  # Convert to datetime
    df[col] = df[col].dt.tz_localize(gmt_tz, ambiguous="NaT", nonexistent="NaT")  # Assign GMT
    df[col] = df[col].dt.tz_convert(pacific_tz)  # Convert to PST/PDT

df['connectionTime'] = pd.to_datetime(df['connectionTime'])
df['connectionTime'] = df['connectionTime'].dt.hour + df['connectionTime'].dt.minute / 60 + df['connectionTime'].dt.second / 3600

df["userInputs/0/modifiedAt"] = pd.to_datetime(df['userInputs/0/modifiedAt'])
df["userInputs/0/modifiedAt"] = df['userInputs/0/modifiedAt'].dt.hour + df['userInputs/0/modifiedAt'].dt.minute / 60 + df['userInputs/0/modifiedAt'].dt.second / 3600

df['userInputs/0/requestedDeparture'] = pd.to_datetime(df['userInputs/0/requestedDeparture'])
df['userInputs/0/requestedDeparture'] = df['userInputs/0/requestedDeparture'].dt.hour + df['userInputs/0/requestedDeparture'].dt.minute / 60 + df['userInputs/0/requestedDeparture'].dt.second / 3600

df['disconnectTime'] = pd.to_datetime(df['disconnectTime'])
df['disconnectTime'] = df['disconnectTime'].dt.hour + df['disconnectTime'].dt.minute / 60 + df['disconnectTime'].dt.second / 3600

df['userInputs/0/paymentRequired'] = df['userInputs/0/paymentRequired'].replace({True: 1, False: 0})

df['spaceID'] = df['spaceID'].str.replace('CA-', '')

df = df.apply(pd.to_numeric, errors='coerce') 
display(df)

df.to_csv('caltech_clean.csv')
```
